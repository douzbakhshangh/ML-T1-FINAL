```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
model = LinearRegression(normalize=True)
x = np.random.rand(5,1)*20
y = 2.35 * x - 3.12
y = y + np.random.normal(scale=3,size=y.shape)
y = y.reshape(-1, 1)
x = x.reshape(-1, 1)
model.fit(x, y)
s1=model.score(x, y)
print(s1)
xline = np.arange(0, 20, 0.1)
xline = xline.reshape(-1, 1)
yline=model.predict(xline)
plt.xlabel('x-axis',fontname='Comic Sans MS', fontsize=17,color='aqua')
plt.ylabel('y-axis',fontname='Comic Sans MS', fontsize=17,color='aqua')
plt.title('out put',fontname='Comic Sans MS', fontsize=20,color='navy')
plt.plot(xline,yline,'m--',label='Part1\nk = 1 & Number = 5')
plt.plot(x,y,'kp', markersize=8)
plt.grid()
plt.legend(loc=2, borderpad=1.5,facecolor='lightyellow',edgecolor='aqua')
plt.savefig('exercise1.png', dpi=300)
```

```python
MSE = []
MSE101 = []
MMSSEE1 = []
MMSSEE2 = []
MMSSEE3 = []
h = model.predict(x)
mse = np.mean((y - h) ** 2) / 2
MSE.append(mse)
MSE101=np.mean(MSE)
MMSSEE1.append(MSE101)
print('MSE101=',MSE101)
print('MMSSEE1=',MMSSEE1)
```
